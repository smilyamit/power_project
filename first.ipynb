```python
# Create figure and plot space
from datetime import date, timedelta, datetime, time
from matplotlib.dates import DayLocator, HourLocator,DateFormatter

fig, ax = plt.subplots(figsize=(10, 7))

# Add x-axis and y-axis
#ax.scatter(dt,y,color='purple')
ax.plot(df[0], df[2], color = 'green')

# set a locator, as well as a formatter
ax.xaxis.set_major_locator(HourLocator(interval = 10))
ax.xaxis.set_major_formatter(DateFormatter("%H"))



#optionally rotate the labels and make more space for them
fig.autofmt_xdate()

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Power Range",
       title="Plot of Power")

plt.show()
plt.savefig('power.jpg')
```
